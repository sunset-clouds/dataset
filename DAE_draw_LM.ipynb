```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

fs=30
ms=30
figsize=(12,10)

##figure 1
beta =[10,20,30,40,50,60,70,80,90]
nll = [329.2,327.5,326.4,325.9,326.1,325.1,325.6,325.7,326.7]
ppl = [30.14,29.61,29.23,29.14,29.20,28.89,29.05,29.08,29.38]

fig = plt.figure(figsize=figsize)
ax1 = fig.add_subplot(111)
line1 = ax1.plot(beta, nll,'-r^',markersize=ms, lw=4, label='NLL')
plt.grid(axis="y")
x_ticks = beta
x_labels = beta
plt.xticks(x_ticks, x_labels, fontsize=fs)
ax1.set_ylabel('NLL',fontsize=fs)
#ax1.set_xlabel('Beta',fontsize=fs)
y1_ticks = np.arange(325, 331, 1)
plt.yticks(y1_ticks, fontsize=fs)

ax2 = ax1.twinx()
line2 = ax2.plot(beta, ppl, '-.bv',markersize=ms, lw=4, label='PPL')
y2_ticks = np.arange(28.5, 31.5, 0.5)
plt.yticks(y2_ticks, fontsize=fs)
ax2.set_ylabel('PPL',fontsize=fs)

lns = line1+line2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs, loc=0,fontsize=ms)


#plt.savefig('Beta.png',bbox_inches = 'tight')
plt.savefig('Beta.pdf',bbox_inches = 'tight')
plt.show()
```
